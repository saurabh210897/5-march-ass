```python
# Glyphs in Bokeh are the visual markers that represent data on a plot, such as circles, squares, lines, bars, or patches. 
# Bokeh provides a wide range of glyphs that can be used to create various types of visualizations.

# To add glyphs to a Bokeh plot, you need to create a ColumnDataSource object that holds the data to be plotted, and then create a glyph object, 
# specifying the type of glyph and its properties, such as size, color, and alpha. Finally, you add the glyph to the plot using the add_glyph method of the plot object.

# Here's an example of how to add circles and lines as glyphs to a Bokeh plot:
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Create some sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [5, 4, 3, 2, 1]

# Create a ColumnDataSource object
source = ColumnDataSource(data=dict(x=x, y1=y1, y2=y2))

# Create a figure object
p = figure(width=400, height=400)

# Add circles and lines as glyphs
p.circle(x='x', y='y1', size=10, color='blue', alpha=0.5, source=source)
p.line(x='x', y='y2', line_width=2, color='red', alpha=0.8, source=source)

# Show the plot
show(p)
```

```python
# Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in 
# real time?
```

```python
# Bokeh is a Python library used for creating interactive data visualizations in web browsers. 
# Bokeh provides a server framework called the Bokeh Server, which allows for the creation of interactive plots that can be updated in real-time.

# The Bokeh Server enables the creation of applications that run on a server and can be accessed by multiple users through their web browsers. 
# The server maintains a connection with the browser, which allows for updates to be pushed from the server to the browser in real-time.

# To create an interactive plot using the Bokeh Server, you first need to create a Bokeh plot as you normally would. 
# You can then add interactive elements to the plot, such as sliders or dropdown menus, that allow users to change the data being displayed.

# Once your plot is ready, you can create a Bokeh application by defining a Python function that generates the plot and handles any user interactions. 
# This function should use the Bokeh Server's document object to update the plot and push any changes to the browser.

# Finally, you can start the Bokeh Server and run your application by executing the command bokeh serve myapp.py in a terminal window, where myapp.py is 
# the Python file containing your application code.

# Once your application is running, users can access it by visiting the appropriate URL in their web browser. 
# Any changes they make to the interactive elements of the plot will be sent back to the server and used to update the plot in real-time.
```
